# Conic Curves

Conic curves are the curves formed by the intersection of a plane and a cone. These curves are solutions to second-degree polynomial equations in two variables and include ellipses, parabolas, and hyperbolas, each with unique properties.

## 1. **General Equation of a Conic Curve**

The general equation for a conic section in Cartesian coordinates \( (x, y) \) is:

$$
Ax^2 + Bxy + Cy^2 + Dx + Ey + F = 0
$$

Where:
- \( A \), \( B \), \( C \), \( D \), \( E \), and \( F \) are constants.
- The nature of the conic (circle, ellipse, parabola, or hyperbola) depends on these constants.

### Classification of Conics:
- **Circle**: \( A = C \) and \( B = 0 \)
- **Ellipse**: \( A \neq C \), \( B = 0 \), and \( AC > 0 \)
- **Parabola**: \( B^2 - 4AC = 0 \)
- **Hyperbola**: \( B^2 - 4AC > 0 \)

## 2. **Standard Forms of Conic Equations**

### a. **Circle**

The equation of a circle with center \( (h, k) \) and radius \( r \) is:

$$
(x - h)^2 + (y - k)^2 = r^2
$$

- **Center**: \( (h, k) \)
- **Radius**: \( r \)

### b. **Ellipse**

The equation of an ellipse with center \( (h, k) \), semi-major axis \( a \), and semi-minor axis \( b \) is:

$$
\frac{(x - h)^2}{a^2} + \frac{(y - k)^2}{b^2} = 1
$$

If \( a = b \), the ellipse becomes a circle.

### c. **Parabola**

The equation of a parabola can be written as:

$$
y - k = a(x - h)^2
$$

or

$$
x - h = a(y - k)^2
$$

Where:
- \( (h, k) \) is the vertex of the parabola,
- \( a \) controls the shape and orientation:
  - If \( a > 0 \), the parabola opens upwards (or to the right for the second form),
  - If \( a < 0 \), it opens downwards (or to the left for the second form).

### d. **Hyperbola**

The equation of a hyperbola with horizontal and vertical transverse axes is:

$$
\frac{(x - h)^2}{a^2} - \frac{(y - k)^2}{b^2} = 1
$$

Where:
- \( (h, k) \) is the center of the hyperbola,
- \( a \) is the distance from the center to each vertex along the transverse axis,
- \( b \) is the distance from the center to each vertex along the conjugate axis.

## 3. **Properties of Conic Curves**

### a. **Circle**
- All points are equidistant from the center.
- The radius is constant.

### b. **Ellipse**
- The sum of the distances from any point on the ellipse to the two foci is constant.
- The foci are located along the major axis.

### c. **Parabola**
- A parabola has a single focus, and every point on the curve is equidistant from the focus and the directrix.

### d. **Hyperbola**
- The difference of the distances from any point on the hyperbola to the two foci is constant.

## 4. **Equations of Conic Sections in Polar Coordinates**

In polar coordinates \( (r, \theta) \), the equations of conic curves are:

### a. **Circle**
For a circle centered at the origin with radius \( R \):

$$
r = R
$$

### b. **Ellipse**

The equation of an ellipse in polar coordinates is:

$$
r = \frac{l}{1 + e \cos \theta}
$$

Where:
- \( l \) is the semi-latus rectum,
- \( e \) is the eccentricity.

### c. **Parabola**

The equation of a parabola in polar coordinates is:

$$
r = \frac{l}{1 + \cos \theta}
$$

### d. **Hyperbola**

The equation of a hyperbola in polar coordinates is:

$$
r = \frac{l}{1 - e \cos \theta}
$$

## 5. **Eccentricity of Conics**

Eccentricity is a measure of how "stretched" a conic is. It is given by:

$$
e = \frac{c}{a}
$$

Where:
- \( c \) is the distance from the center to the focus,
- \( a \) is the distance from the center to a vertex.

For different conics:
- **Circle**: \( e = 0 \)
- **Ellipse**: \( 0 < e < 1 \)
- **Parabola**: \( e = 1 \)
- **Hyperbola**: \( e > 1 \)

## 6. **Applications of Conic Curves**

- **Ellipses**: Used in planetary motion (Kepler’s laws), acoustics, and optics (elliptical reflectors).
- **Parabolas**: Used in satellite dishes, reflectors, and projectile motion.
- **Hyperbolas**: Used in navigation systems and modeling paths under specific gravitational conditions.

## 7. **Conic Sections as Degenerate Cases**

When the coefficients of the general equation are special values, degenerate conics arise:
- A **degenerate ellipse** is a point.
- A **degenerate hyperbola** is a pair of intersecting lines.
- A **degenerate parabola** is a line.

## Conclusion

Conic sections are essential in geometry and appear in various applications across physics, engineering, and astronomy. Understanding their equations, properties, and polar representations deepens our understanding of their significance and utility.

# Solutions:

### **1. Equation of a Circle**


The standard equation of a circle with center at \( (h, k) \) and radius \( r \) is:

$$
(x - h)^2 + (y - k)^2 = r^2
$$

Given the center at \( A(1, 2) \) and radius \( r = 3 \), the equation becomes:

$$
(x - 1)^2 + (y - 2)^2 = 9
$$

This represents the equation of the circle.

```python
import numpy as np
import matplotlib.pyplot as plt

# **1. Equation of a Circle**
# Circle with center A(1, 2) and radius r = 3
center = (1, 2)
radius = 3

theta = np.linspace(0, 2 * np.pi, 300)
x = center[0] + radius * np.cos(theta)
y = center[1] + radius * np.sin(theta)

plt.figure(figsize=(6,6))
plt.plot(x, y, label="Circle: $(x-1)^2 + (y-2)^2 = 9$")
plt.scatter(center[0], center[1], color='red', zorder=5, label="Center (1, 2)")
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Equation of a Circle")
plt.legend()
plt.grid(True)
plt.show()
```

### **2. Equation of a Parabola**

We are given a parabola that intersects the $x$-axis at $x = 2$ and $x = 4$, and passes through the point $(3, 1)$.

The general form of the parabola equation is:

$$
y = a(x - p)(x - q)
$$

Where $(p, 0)$ and $(q, 0)$ are the points where the parabola intersects the $x$-axis.

In this case, the intersections are at $x = 2$ and $x = 4$, so the equation becomes:

$$
y = a(x - 2)(x - 4)
$$

We are also told that the parabola passes through the point $(3, 1)$. Substituting $x = 3$ and $y = 1$ into the equation:

$$
1 = a(3 - 2)(3 - 4)
$$

Simplifying the equation:

$$
1 = a(1)(-1) \quad \Rightarrow \quad a = -1
$$

Thus, the equation of the parabola is:

$$
y = -(x - 2)(x - 4)
$$

Expanding the equation:

$$
y = -(x^2 - 6x + 8) \quad \Rightarrow \quad y = -x^2 + 6x - 8
$$

This is the equation of the parabola.

```python
import numpy as np
import matplotlib.pyplot as plt
# **2. Equation of a Parabola**
# Parabola: y = -(x-2)(x-4)
x_vals = np.linspace(1, 5, 400)
y_vals = -(x_vals - 2) * (x_vals - 4)

plt.plot(x_vals, y_vals, label=r'$y = -(x-2)(x-4)$')
plt.scatter([3], [1], color='red', zorder=5, label="Point (3, 1)")
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.title("Equation of a Parabola")
plt.legend()
plt.grid(True)
plt.show()
```

### **3. Center of the Ellipse**

We are given the equation of an ellipse:

$$
x^2 + 4y^2 - 4x - 16y + 16 = 0
$$

To find the center, we need to rewrite this equation in standard form by completing the square for both $x$ and $y$ terms.

#### Completing the square for the $x$-terms:
$$
x^2 - 4x \quad \text{(add and subtract 4)} \quad (x - 2)^2 - 4
$$

#### Completing the square for the $y$-terms:
$$
4y^2 - 16y \quad \text{(factor out the 4)} \quad 4(y^2 - 4y) \quad \text{(add and subtract 4 inside the parentheses)} \quad 4((y - 2)^2 - 4)
$$

Now substitute these into the original equation:

$$
(x - 2)^2 - 4 + 4((y - 2)^2 - 4) + 16 = 0
$$

Simplifying:

$$
(x - 2)^2 - 4 + 4(y - 2)^2 - 16 + 16 = 0 \quad \Rightarrow \quad (x - 2)^2 + 4(y - 2)^2 = 4
$$

The equation is now in standard form, and we can identify that the center of the ellipse is at:

$$
(2, 2)
$$

```python
# **3. Center of the Ellipse**
# Ellipse: (x-2)^2 + 4(y-2)^2 = 4
theta = np.linspace(0, 2 * np.pi, 300)
x_vals = 2 + 2 * np.cos(theta)
y_vals = 2 + np.sin(theta)

plt.plot(x_vals, y_vals, label=r'$(x-2)^2 + 4(y-2)^2 = 4$')
plt.scatter(2, 2, color='red', zorder=5, label="Center (2, 2)")
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Center of the Ellipse")
plt.legend()
plt.grid(True)
plt.show()
```

### **4. Slope of the Tangent Line to a Circle**

We are tasked with finding the slope ($m>0$) of the line $y = mx - 5$ that is tangent to the circle given by the equation:

$$
x^2 + y^2 = 1
$$

For the line to be tangent to the circle, the distance from the center of the circle, $(0, 0)$, to the line must equal the radius of the circle, which is 1.

The general formula for the distance from a point $(x_1, y_1)$ to a line $Ax + By + C = 0$ is:

$$
\text{Distance} = \frac{|Ax_1 + By_1 + C|}{\sqrt{A^2 + B^2}}
$$

The equation of the line is $y = mx - 5$, which can be rewritten as:

$$
mx - y - 5 = 0
$$

Thus, the constants are $A = m$, $B = -1$, and $C = -5$. The distance from the center $(0, 0)$ to the line is:

$$
\frac{|m(0) - 1(0) - 5|}{\sqrt{m^2 + (-1)^2}} = \frac{5}{\sqrt{m^2 + 1}}
$$

Since the line is tangent to the circle, this distance must be equal to the radius, which is 1. Therefore, we set the equation equal to 1:

$$
\frac{5}{\sqrt{m^2 + 1}} = 1
$$

Squaring both sides:

$$
\frac{25}{m^2 + 1} = 1 \quad \Rightarrow \quad m^2 + 1 = 25 \quad \Rightarrow \quad m^2 = 24 \quad \Rightarrow \quad m = \sqrt{24} = 2\sqrt{6}
$$

Thus, the slope of the tangent line is:

$$
m = 2\sqrt{6}
$$

```python
# **4. Slope of the Tangent Line to a Circle**
# Circle: x^2 + y^2 = 1, tangent line: y = 2√6 x - 5
theta = np.linspace(0, 2 * np.pi, 300)
x_vals = np.cos(theta)
y_vals = np.sin(theta)

# Tangent line: y = 2√6 x - 5
x_line = np.linspace(-1, 1, 300)
y_line = 2 * np.sqrt(6) * x_line - 5

plt.plot(x_vals, y_vals, label=r'$x^2 + y^2 = 1$ (Circle)')
plt.plot(x_line, y_line, label=r'$y = 2\sqrt{6}x - 5$ (Tangent Line)', linestyle='--')
plt.scatter([0], [1], color='red', label="Point on Circle (0, 1)")
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Tangent Line to the Circle")
plt.legend()
plt.grid(True)
plt.show()
```

### **5. Intersection Points of Hyperbola and Ellipse**

We are tasked with finding the intersection points of the hyperbola:

$$
x^2 - y^2 = 1
$$

and the ellipse:

$$
x^2 + 4y^2 = 6
$$

To find the intersection, we solve the two equations simultaneously. Start by solving one of the equations for $x^2$ and substituting it into the other equation.

From the hyperbola equation:

$$
x^2 = y^2 + 1
$$

Substitute this expression for $x^2$ into the ellipse equation:

$$
(y^2 + 1) + 4y^2 = 6
$$

Simplify:

$$
5y^2 + 1 = 6 \quad \Rightarrow \quad 5y^2 = 5 \quad \Rightarrow \quad y^2 = 1
$$

Thus, we have two possible values for $y$:

$$
y = \pm 1
$$

For each $y$ value, we substitute back into the equation $x^2 = y^2 + 1$:

- For $y = 1$:

$$
x^2 = 1 + 1 = 2 \quad \Rightarrow \quad x = \pm \sqrt{2}
$$

- For $y = -1$:

$$
x^2 = (-1)^2 + 1 = 2 \quad \Rightarrow \quad x = \pm \sqrt{2}
$$

Thus, the intersection points are:

$$
(\sqrt{2}, 1), \quad (-\sqrt{2}, 1), \quad (\sqrt{2}, -1), \quad (-\sqrt{2}, -1)
$$

```python
# **5. Intersection Points of Hyperbola and Ellipse**
# Hyperbola: x^2 - y^2 = 1, Ellipse: x^2 + 4y^2 = 6
x_vals = np.linspace(-3, 3, 400)
y_vals_hyperbola_pos = np.sqrt(x_vals**2 - 1)
y_vals_hyperbola_neg = -np.sqrt(x_vals**2 - 1)

# Ellipse
y_vals_ellipse_pos = np.sqrt((6 - x_vals**2) / 4)
y_vals_ellipse_neg = -np.sqrt((6 - x_vals**2) / 4)

plt.plot(x_vals, y_vals_hyperbola_pos, label=r'$x^2 - y^2 = 1$ (Hyperbola)', color='green')
plt.plot(x_vals, y_vals_hyperbola_neg, color='green')
plt.plot(x_vals, y_vals_ellipse_pos, label=r'$x^2 + 4y^2 = 6$ (Ellipse)', color='blue')
plt.plot(x_vals, y_vals_ellipse_neg, color='blue')

plt.scatter([np.sqrt(2), -np.sqrt(2)], [1, 1], color='red', zorder=5, label="Intersection Points")
plt.scatter([np.sqrt(2), -np.sqrt(2)], [-1, -1], color='red')
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Intersection of Hyperbola and Ellipse")
plt.legend()
plt.grid(True)
plt.show()
```

### **6. Distance Between Branches of the Hyperbola**

For the hyperbola:

$$
x^2 - y^2 = 1
$$

The branches of the hyperbola are symmetric with respect to the $x$-axis, and the distance between the branches at any point is determined by the $y$-coordinate. The branches approach the $x$-axis as $|x|$ becomes large.

To find the distance between the branches at $y = 0$, we substitute $y = 0$ into the equation of the hyperbola:

$$
x^2 - 0^2 = 1 \quad \Rightarrow \quad x^2 = 1 \quad \Rightarrow \quad x = \pm 1
$$

The distance between the branches is the distance between the points where the hyperbola intersects the $x$-axis, which are at $x = 1$ and $x = -1$. Therefore, the distance is:

$$
\text{Distance} = 1 - (-1) = 2
$$

Thus, the distance between the branches of the hyperbola at $y = 0$ is 2.

```python
# **6. Distance Between Branches of the Hyperbola**
# Hyperbola: x^2 - y^2 = 1
x_vals = np.linspace(-3, 3, 400)
y_vals_hyperbola_pos = np.sqrt(x_vals**2 - 1)
y_vals_hyperbola_neg = -np.sqrt(x_vals**2 - 1)

plt.plot(x_vals, y_vals_hyperbola_pos, label=r'$x^2 - y^2 = 1$ (Hyperbola)', color='green')
plt.plot(x_vals, y_vals_hyperbola_neg, color='green')
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Distance Between Branches of the Hyperbola")
plt.legend()
plt.grid(True)
plt.show()
```
