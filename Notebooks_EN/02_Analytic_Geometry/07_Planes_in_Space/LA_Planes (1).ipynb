### Planes in Space
A plane in space is an infinitely extending flat surface. It is defined by points and vectors, with its orientation determined by a **normal vector**, which is perpendicular to the plane.

---

### Vector Equation of a Plane
The vector equation of a plane passing through the point $ \mathbf{P}_0(x_0, y_0, z_0) $ with a normal vector $ \mathbf{n} = (a, b, c) $ is expressed as:

$$
\mathbf{n} \cdot (\mathbf{r} - \mathbf{r}_0) = 0
$$

where:
- $ \mathbf{r} = (x, y, z) $ is the position vector of any point on the plane,
- $ \mathbf{r}_0 = (x_0, y_0, z_0) $ is the position vector of a given point,
- $ \mathbf{n} \cdot (\mathbf{r} - \mathbf{r}_0) $ is the dot product.

By expanding this, the equation becomes:

$$
a(x - x_0) + b(y - y_0) + c(z - z_0) = 0
$$

---

### General Equation of a Plane
The general equation of a plane in space can be written as:

$$
ax + by + cz + d = 0
$$

where:
- \( a, b, c \) are the components of the normal vector $ \mathbf{n} $,
- $ d $ is the constant term.

---

### Plane Equation Passing Through Three Points
If three points $ \mathbf{P}_1(x_1, y_1, z_1) $, $ \mathbf{P}_2(x_2, y_2, z_2) $, and $ \mathbf{P}_3(x_3, y_3, z_3) $ define the plane, the normal vector $ \mathbf{n} $ can be calculated as:

$$
\mathbf{n} = (\mathbf{P}_2 - \mathbf{P}_1) \times (\mathbf{P}_3 - \mathbf{P}_1)
$$

Using this, the plane equation becomes:

$$
a(x - x_1) + b(y - y_1) + c(z - z_1) = 0
$$

---

### Plane Equation in Intercept Form
For a plane that intersects the coordinate axes at $ (a, 0, 0) $, $ (0, b, 0) $, and $ (0, 0, c) $, the equation can be written in intercept form as:

$$
\frac{x}{a} + \frac{y}{b} + \frac{z}{c} = 1
$$

---

### Special Cases of Plane Positioning
1. **Parallel to a Coordinate Plane**:
   - A plane parallel to the $ XY $-plane: $ z = k $,
   - A plane parallel to the $ XZ $-plane: $ y = k $,
   - A plane parallel to the $ YZ $-plane: $ x = k $.
   
2. **Coinciding with a Coordinate Plane**:
   - A plane coincident with the $ XY $-plane: $ z = 0 $,
   - A plane coincident with the $ XZ $-plane: $ y = 0 $,
   - A plane coincident with the $ YZ $-plane: $ x = 0 $.

---

### Angle Between Two Planes
The angle $ \theta $ between two planes with normal vectors $ \mathbf{n}_1 = (a_1, b_1, c_1) $ and $ \mathbf{n}_2 = (a_2, b_2, c_2) $ is calculated as:

$$
\cos \theta = \frac{\mathbf{n}_1 \cdot \mathbf{n}_2}{\|\mathbf{n}_1\| \|\mathbf{n}_2\|}
$$

---

### Conditions for Parallelism and Perpendicularity of Two Planes
1. **Parallel Planes**:
   Two planes are parallel if their normal vectors are scalar multiples:
   $$
   \mathbf{n}_1 \parallel \mathbf{n}_2 \quad \Leftrightarrow \quad \frac{a_1}{a_2} = \frac{b_1}{b_2} = \frac{c_1}{c_2}
   $$

2. **Perpendicular Planes**:
   Two planes are perpendicular if the dot product of their normal vectors is zero:
   $$
   \mathbf{n}_1 \cdot \mathbf{n}_2 = 0
   $$

## Lines in Space

### Types of Line Equations in Space

1. **Parametric Equation**:
   A line passing through the point $ \mathbf{P}_0(x_0, y_0, z_0) $ in the direction of vector $ \mathbf{v} = (a, b, c) $ is given by:

  $$
   x = x_0 + at, \quad y = y_0 + bt, \quad z = z_0 + ct
  $$

2. **Symmetric Equation**:
   By eliminating $ t $ from the parametric equations, we obtain the symmetric equation:

   $$
   \frac{x - x_0}{a} = \frac{y - y_0}{b} = \frac{z - z_0}{c}
   $$

---

### Relative Positions of a Line and a Plane
1. **Line Lies on the Plane**:
   Substitute the line's parametric equations into the plane's equation. If the equation holds for all $ t $, the line lies on the plane.

2. **Line is Parallel to the Plane**:
   If $ \mathbf{v} \cdot \mathbf{n} = 0 $, the line is parallel to the plane, where $ \mathbf{v} $ is the direction vector of the line and $ \mathbf{n} $ is the normal vector of the plane.

3. **Line Intersects the Plane**:
   Substitute the parametric equations of the line into the equation of the plane and solve for $ t $ to find the intersection.

---

### Relative Positions of Two Lines
1. **Parallel Lines**:
   Two lines are parallel if their direction vectors $ \mathbf{v}_1 $ and $ \mathbf{v}_2 $ are scalar multiples:
   $$
   \mathbf{v}_1 \parallel \mathbf{v}_2
   $$

2. **Intersecting Lines**:
   Lines intersect if there is a common solution to their parametric equations.

3. **Skew Lines**:
   Lines are skew if they are neither parallel nor intersecting.

---

### Angle Between a Line and a Plane
The angle $ \theta $ between a line with direction vector $ \mathbf{v} $ and a plane with normal vector $ \mathbf{n} $ is given by:

$$
\sin \theta = \frac{\mathbf{v} \cdot \mathbf{n}}{\|\mathbf{v}\| \|\mathbf{n}\|}
$$

Alternatively, we can express it as:

$$
\cos \theta = \sqrt{1 - \sin^2 \theta}
$$

### **Solution for Plane through Points $A(1, 2, 3)$, $B(3, 4, 5)$, and $C(2, 1, 4)$**

1. **Compute vectors $\vec{AB}$ and $\vec{AC}$:**
   \[
   \vec{AB} = B - A = (3 - 1, 4 - 2, 5 - 3) = (2, 2, 2)
   \]
   \[
   \vec{AC} = C - A = (2 - 1, 1 - 2, 4 - 3) = (1, -1, 1)
   \]

2. **Compute the cross product $\vec{n} = \vec{AB} \times \vec{AC}$:**
   \[
   \vec{n} = \begin{vmatrix}
   \mathbf{i} & \mathbf{j} & \mathbf{k} \\
   2 & 2 & 2 \\
   1 & -1 & 1 \\
   \end{vmatrix}
   = \mathbf{i}(2 \cdot 1 - (-2)) - \mathbf{j}(2 \cdot 1 - 2) + \mathbf{k}(2 \cdot (-1) - 2)
   \]
   \[
   \vec{n} = (4, 0, -4)
   \]

3. **Write the equation of the plane:**
   Using the point $A(1, 2, 3)$ and the normal vector $\vec{n} = (4, 0, -4)$, the plane equation is:
   \[
   4(x - 1) + 0(y - 2) - 4(z - 3) = 0
   \]
   Simplify:
   \[
   4(x - 1) - 4(z - 3) = 0
   \]
   \[
   x - z = -2
   \]
   Hence, the equation of the plane is:
   \[
   x - z + 2 = 0
   \]


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Helper functions
def plot_plane(ax, A, B, C, label):
    """Plots a plane given by Ax + By + Cz = D."""
    x = np.linspace(-5, 5, 10)
    y = np.linspace(-5, 5, 10)
    X, Y = np.meshgrid(x, y)
    Z = (D - A * X - B * Y) / C
    ax.plot_surface(X, Y, Z, alpha=0.5, label=label)

def plot_point(ax, point, color='red', label=None):
    """Plots a single point in 3D."""
    ax.scatter(point[0], point[1], point[2], color=color, label=label)

def plot_vector(ax, origin, vector, color='blue', label=None):
    """Plots a vector from the origin."""
    ax.quiver(
        origin[0], origin[1], origin[2],
        vector[0], vector[1], vector[2],
        color=color, label=label
    )

# Problem 1: Plane through points A, B, C
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
A, B, C = np.array([1, 2, 3]), np.array([3, 4, 5]), np.array([2, 1, 4])

# Compute normal vector
AB = B - A
AC = C - A
normal = np.cross(AB, AC)

# Plane equation: normal[0]*x + normal[1]*y + normal[2]*z = D
D = np.dot(normal, A)

# Plot plane
plot_plane(ax, normal[0], normal[1], normal[2], label="Plane through A, B, C")
plot_point(ax, A, label="Point A")
plot_point(ax, B, label="Point B")
plot_point(ax, C, label="Point C")
plt.title("Plane Through Points A, B, C")
plt.show()
```

### **Solution for Plane Passing Through Point $A(1, 2, 3)$ and Parallel to the Plane $2x + 3y + 4z = 5$**

Planes that are parallel share the same normal vector. The given plane equation $2x + 3y + 4z = 5$ has the normal vector $\vec{n} = (2, 3, 4)$.

Since the required plane passes through point $A(1, 2, 3)$ and is parallel to the given plane, the equation of the new plane will have the same normal vector and satisfy the condition of passing through $A$.

1. **Write the equation of the plane:**
   Using the point $A(1, 2, 3)$ and the normal vector $\vec{n} = (2, 3, 4)$, the equation of the plane is:
   \[
   2(x - 1) + 3(y - 2) + 4(z - 3) = 0
   \]

2. **Simplify the equation:**
   \[
   2(x - 1) + 3(y - 2) + 4(z - 3) = 0
   \]
   Expanding this:
   \[
   2x - 2 + 3y - 6 + 4z - 12 = 0
   \]
   Combine like terms:
   \[
   2x + 3y + 4z - 20 = 0
   \]

Thus, the equation of the plane passing through $A(1, 2, 3)$ and parallel to the plane $2x + 3y + 4z = 5$ is:
\[
2x + 3y + 4z - 20 = 0
\]

```python
# Problem 2: Plane through A and parallel to another plane
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
point = np.array([1, 2, 3])
normal = np.array([2, 3, 4])
D = np.dot(normal, point)
plot_plane(ax, normal[0], normal[1], normal[2], label="Parallel Plane")
plot_point(ax, point, label="Point A")
plt.title("Plane Through A Parallel to Another Plane")
plt.show()
```

## **4. Find the Line of Intersection of Two Planes $2x + 3y + 4z = 5$ and $3x + 4y + 2z = 6$.**

### Solution:
1. Solve the system:
   $$
   \begin{aligned}
   & 2x + 3y + 4z = 5 \\
   & 3x + 4y + 2z = 6
   \end{aligned}
   $$

2. Parameterize one variable, say $z = t$:
   Substitute $z = t$ into both equations:
   $$
   2x + 3y + 4t = 5 \\
   3x + 4y + 2t = 6
   $$

3. Solve for $x$ and $y$ in terms of $t$:
   Simplify to find parametric equations:
   $$
   x = \dots, \quad y = \dots, \quad z = t
   $$

```python
# Problem 4: Line of intersection of two planes
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
# Planes: 2x + 3y + 4z = 5 and 3x + 4y + 2z = 6
A1, B1, C1, D1 = 2, 3, 4, 5
A2, B2, C2, D2 = 3, 4, 2, 6
# Cross product of normals gives direction vector
dir_vector = np.cross([A1, B1, C1], [A2, B2, C2])
# Solve for a point on the line
t = np.linspace(-10, 10, 100)
# Parametrize the line
x_line = t * dir_vector[0]
y_line = t * dir_vector[1]
z_line = t * dir_vector[2]
ax.plot(x_line, y_line, z_line, label="Intersection Line", color="green")
plt.title("Intersection Line of Two Planes")
plt.show()
```

### **Solution: Equation of the Plane Passing Through $A(1, 2, 3)$ and Parallel to Vectors $\vec{v_1} = [1, 0, 1]$ and $\vec{v_2} = [0, 1, -1]$**

1. **Find the normal vector $\vec{n}$:**
   The normal vector to the plane is given by the cross product of the vectors $\vec{v_1}$ and $\vec{v_2}$. We compute the cross product as:
   $$
   \vec{n} = \vec{v_1} \times \vec{v_2} = \begin{vmatrix}
   \mathbf{i} & \mathbf{j} & \mathbf{k} \\
   1 & 0 & 1 \\
   0 & 1 & -1 \\
   \end{vmatrix}
   = \mathbf{i}(0 \cdot (-1) - 1 \cdot 1) - \mathbf{j}(1 \cdot (-1) - 0 \cdot 1) + \mathbf{k}(1 \cdot 1 - 0 \cdot 0)
   $$
   $$
   \vec{n} = (-1, 1, 1)
   $$

2. **Write the equation of the plane:**
   Using the point $A(1, 2, 3)$ and the normal vector $\vec{n} = (-1, 1, 1)$, the equation of the plane is:
   $$
   -1(x - 1) + 1(y - 2) + 1(z - 3) = 0
   $$

3. **Simplify the equation:**
   Expanding and simplifying the terms:
   $$
   -x + 1 + y - 2 + z - 3 = 0
   $$
   $$
   -x + y + z - 4 = 0
   $$

Thus, the equation of the plane is:
$$
-x + y + z - 4 = 0
$$

```python
# Problem 5: Plane through A parallel to two vectors
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
v1 = np.array([1, 0, 1])
v2 = np.array([0, 1, -1])
normal = np.cross(v1, v2)
D = np.dot(normal, A)
plot_plane(ax, normal[0], normal[1], normal[2], label="Plane")
plot_vector(ax, A, v1, color="blue", label="v1")
plot_vector(ax, A, v2, color="green", label="v2")
plt.title("Plane Through A Parallel to Two Vectors")
plt.show()
```

## **9. Angle Between Planes $x + y + z = 1$ and $x = 0$.**

### Solution:
The angle $\theta$ between two planes is given by:
$$
\cos\theta = \frac{|\vec{n_1} \cdot \vec{n_2}|}{\|\vec{n_1}\| \|\vec{n_2}\|}
$$
For $x + y + z = 1$, $\vec{n_1} = (1, 1, 1)$.
For $x = 0$, $\vec{n_2} = (1, 0, 0)$.
$$
\cos\theta = \frac{|1 \cdot 1 + 1 \cdot 0 + 1 \cdot 0|}{\sqrt{1^2 + 1^2 + 1^2} \sqrt{1^2}} = \frac{1}{\sqrt{3}}
$$
$$
\theta = \arccos\left(\frac{1}{\sqrt{3}}\right)
$$

```python
# Problem 9: Angle between planes
# Planes: x + y + z = 1 and x = 0 (yz-plane)
normal1 = np.array([1, 1, 1])
normal2 = np.array([1, 0, 0])

# Angle formula: cos(theta) = (n1 · n2) / (|n1| * |n2|)
cos_theta = np.dot(normal1, normal2) / (np.linalg.norm(normal1) * np.linalg.norm(normal2))
angle = np.arccos(cos_theta)
print(f"Angle between planes: {np.degrees(angle):.2f} degrees")

# Visualization
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
plot_plane(ax, normal1[0], normal1[1], normal1[2], label="Plane 1")
plot_plane(ax, normal2[0], normal2[1], normal2[2], label="Plane 2")
plt.title("Angle Between Planes")
plt.show()
```
