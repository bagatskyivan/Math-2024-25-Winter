#### **1. Taylor Series of $f(x) = \cos(x)$ around $x = 0$**

1.1. The general formula for the Taylor series is:
$$
f(x) = f(0) + f'(0)x + \frac{f''(0)x^2}{2!} + \frac{f'''(0)x^3}{3!} + \dots
$$

1.2. Compute the derivatives of $f(x)$:
- $f(x) = \cos(x)$
- $f'(x) = -\sin(x)$
- $f''(x) = -\cos(x)$
- $f'''(x) = \sin(x)$
- $f^{(4)}(x) = \cos(x)$ (repeats every 4 terms)

1.3. Evaluate the derivatives at $x = 0$:
- $f(0) = \cos(0) = 1$
- $f'(0) = -\sin(0) = 0$
- $f''(0) = -\cos(0) = -1$
- $f'''(0) = \sin(0) = 0$
- $f^{(4)}(0) = \cos(0) = 1$

1.4. Substitute into the Taylor series:
$$
f(x) = 1 - \frac{x^2}{2!} + \frac{x^4}{4!} + \mathcal{O}(x^6)
$$

#### **2. Taylor Series of $h(x) = \frac{1}{1 - x}$ around $x = 0$**

2.1. The function $h(x)$ is a geometric series, which can be expanded as:
$$
h(x) = 1 + x + x^2 + x^3 + \dots
$$

2.2. For a Taylor series up to the 4th degree:
$$
h(x) = 1 + x + x^2 + x^3 + \mathcal{O}(x^4)
$$

---

#### **3. Taylor Series of $g(x) = \sin(x)$ around $x = \pi$**

3.1. Let $y = x - \pi$. Then:
$$
g(x) = \sin(x) = \sin(y + \pi)
$$

3.2. Use the trigonometric identity $\sin(y + \pi) = -\sin(y)$:
$$
g(x) = -\sin(y)
$$

3.3. Expand $-\sin(y)$ around $y = 0$:
$$
-\sin(y) = -(y - \frac{y^3}{3!} + \mathcal{O}(y^5))
$$

3.4. Substitute $y = x - \pi$:
$$
g(x) = -(x - \pi) + \frac{(x - \pi)^3}{6} + \mathcal{O}((x - \pi)^5)
$$

#### **4. Tangent Line to $f(x) = e^{\sin(x)}$ at $x_0 = \pi$**

4.1. Compute the derivative of $f(x)$:
$$
f(x) = e^{\sin(x)}
$$
$$
f'(x) = e^{\sin(x)} \cdot \cos(x)
$$

4.2. Evaluate $f'(\pi)$:
$$
f'(\pi) = e^{\sin(\pi)} \cdot \cos(\pi)
$$
$$
f'(\pi) = e^0 \cdot (-1) = -1
$$

4.3. Evaluate $f(\pi)$:
$$
f(\pi) = e^{\sin(\pi)} = e^0 = 1
$$

4.4. Write the tangent line equation:
$$
y = f'(x_0)(x - x_0) + f(x_0)
$$

Substitute $x_0 = \pi$, $f'(x_0) = -1$, and $f(x_0) = 1$:
$$
y = -1(x - \pi) + 1
$$

Simplify:
$$
y = -x + \pi + 1
$$
