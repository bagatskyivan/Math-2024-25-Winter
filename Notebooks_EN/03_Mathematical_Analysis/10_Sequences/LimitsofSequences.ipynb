

## 19. Limits of Sequences

1. Calculate:
   $$\lim_{n \to \infty} \frac{n^2 + 3n}{2n^2 - 2n}$$
   $$\lim_{n \to \infty} \frac{(2n + 3)^3}{n^3 - 1}$$

2. Prove using the squeeze theorem:
   $$\lim_{n \to \infty} \frac{\sin(n)}{n}$$

3. Find the limit of the sequence:
   $$a_n = \left(1 + \frac{1}{n}\right)^n$$


----
### 1. Calculate:

#### $$
$$(a) $$
\lim_{n \to \infty} \frac{n^2 + 3n}{2n^2 - 2n}$$

1. Factor out \(n^2\) from both the numerator and denominator:

$$
\frac{n^2 + 3n}{2n^2 - 2n} = \frac{n^2(1 + \frac{3}{n})}{n^2(2 - \frac{2}{n})}.
$$

2. Simplify the expression:

$$
\frac{1 + \frac{3}{n}}{2 - \frac{2}{n}}.
$$

3. Take the limit as $$n \to \infty$$:

$$
\lim_{n \to \infty} \frac{1 + \frac{3}{n}}{2 - \frac{2}{n}} = \frac{1}{2}.
$$

#### (b) $$\lim_{n \to \infty} \frac{(2n + 3)^3}{n^3 - 1}$$

1. Expand \((2n + 3)^3\):

$$
(2n + 3)^3 = 8n^3 + 12n^2 + 18n + 27.
$$

2. Divide both numerator and denominator by \(n^3\):

$$
\frac{8n^3 + 12n^2 + 18n + 27}{n^3 - 1} = \frac{8 + \frac{12}{n} + \frac{18}{n^2} + \frac{27}{n^3}}{1 - \frac{1}{n^3}}.
$$

3. Take the limit as \(n \to \infty\):

$$
\lim_{n \to \infty} \frac{8 + \frac{12}{n} + \frac{18}{n^2} + \frac{27}{n^3}}{1 - \frac{1}{n^3}} = 8.
$$


-----

### 2. Prove using the squeeze theorem:

$$
\lim_{n \to \infty} \frac{\sin(n)}{n}
$$

1. Use the inequality $$|\sin(n)| \leq 1$$:

$$
-\frac{1}{n} \leq \frac{\sin(n)}{n} \leq \frac{1}{n}.
$$

2. Take the limit of both bounds as $$n \to \infty$$:

$$
\lim_{n \to \infty} -\frac{1}{n} = 0 \quad \text{and} \quad \lim_{n \to \infty} \frac{1}{n} = 0.
$$

3. By the squeeze theorem:

$$
\lim_{n \to \infty} \frac{\sin(n)}{n} = 0.
$$

### 3. Find the limit of the sequence:

$$
a_n = \left(1 + \frac{1}{n}\right)^n
$$

1. Recall the limit definition of \(e\):

$$
\lim_{n \to \infty} \left(1 + \frac{1}{n}\right)^n = e.
$$

2. Conclude:

$$
\lim_{n \to \infty} a_n = e.
$$
