## 1. Compute:

### 1.
$$
\int 1dx = x + C
$$

### 2.
$$
\int (x^2 + 2)dx = \int x^2 dx + \int 2 dx = \frac{x^3}{3} + 2x + C
$$

### 3.
$$
\int 2\sin(x)dx = 2\int \sin(x)dx = -2\cos(x) + C
$$

### 4.
$$
\int \frac{3}{x}dx = 3\int \frac{1}{x}dx = 3\ln|x| + C
$$

### 5.
$$
\int \frac{1}{x^2}dx = \int x^{-2}dx = \frac{x^{-1}}{-1} = -\frac{1}{x} + C
$$

### 6.
$$
\int \left(\frac{1}{3}x^4 - 5\right)dx = \frac{1}{3} \int x^4 dx - 5 \int dx
$$
$$
= \frac{1}{3} \cdot \frac{x^5}{5} - 5x + C = \frac{1}{15}x^5 - 5x + C
$$

### 7.
$$
\int (\sin^2x + \cos^2x)dx = \int 1 dx = x + C
$$

### 8.
$$
\int (5\sin x + 3e^x)dx = 5\int \sin x dx + 3\int e^x dx = -5\cos x + 3e^x + C
$$

### 9.
$$
\int \sqrt[3]{x}dx = \int x^{\frac{1}{3}}dx = \frac{x^{\frac{4}{3}}}{\frac{4}{3}} = \frac{3}{4}x^{\frac{4}{3}} + C
$$

### 10.
$$
\int \sqrt{10x}dx = \sqrt{10} \int x^{\frac{1}{2}}dx = \sqrt{10} \cdot \frac{x^{\frac{3}{2}}}{\frac{3}{2}} = \frac{2\sqrt{10}}{3}x^{\frac{3}{2}} + C
$$

### 11.
$$
\int \cos\left(\frac{5}{2}x + 3\right)dx
$$
Let $ u = \frac{5}{2}x + 3 $, then $ du = \frac{5}{2}dx $, so
$$
\frac{2}{5} \int \cos u \, du = \frac{2}{5} \sin u + C = \frac{2}{5} \sin\left(\frac{5}{2}x + 3\right) + C
$$

### 12.
$$
\int \frac{\cos(\ln(x))}{x}dx = \int \cos(\ln x) d(\ln x) = \sin(\ln x) + C
$$

### 13.
Integration by parts: let $ u = \ln x $, $ dv = x dx $, so $ du = \frac{1}{x}dx $, $ v = \frac{x^2}{2} $.

$$
\int x\ln(x)dx = \frac{x^2}{2} \ln x - \int \frac{x^2}{2} \cdot \frac{1}{x}dx
$$

$$
= \frac{x^2}{2} \ln x - \int \frac{x}{2}dx = \frac{x^2}{2} \ln x - \frac{x^2}{4} + C
$$

### 14.
Integration by parts: let $ u = x $, $ dv = e^x dx $, so $ du = dx $, $ v = e^x $.

$$
\int xe^x dx = x e^x - \int e^x dx = x e^x - e^x + C = (x - 1)e^x + C
$$

## 2. Calculate integrals over the interval $[0, \pi]$ and visualize them in Geogebra:


### **Integral of \( f(x) = 2x + 1 \)**
We compute the definite integral over the interval \([0, $\pi]$\):

$$ I_f = \int_0^\pi (2x + 1) \, dx $$

Computing the antiderivative:

$$ \int (2x + 1) \, dx = x^2 + x $$

Evaluating from \( 0 \) to \( $\pi $\):

$$ I_f = (\pi^2 + \pi) - (0^2 + 0) = \pi^2 + \pi $$

### **Integral of \( g(x) = x^2 \)**
Similarly, for \( g(x) = x^2 \):

$$ I_g = \int_0^\pi x^2 \, dx $$

Computing the antiderivative:

$$ \int x^2 \, dx = \frac{x^3}{3} $$

Evaluating from \( 0$) to $( $\pi$\):

$$ I_g = \frac{\pi^3}{3} - \frac{0^3}{3} = \frac{\pi^3}{3} $$

## 3. Calculate the area of the region bounded by the lines:
### **Step 1: Compute the Area**  
The required area is given by the definite integral:

$$
A = \int_{1}^{2} (x^2 + 1) \,dx
$$

Computing the integral:

$$
\int (x^2 + 1)dx = \frac{x^3}{3} + x
$$

Evaluating from \( x = 1 \) to \( x = 2 \):

$$
A = \left(\frac{2^3}{3} + 2\right) - \left(\frac{1^3}{3} + 1\right)
$$

$$
= \left(\frac{8}{3} + 2\right) - \left(\frac{1}{3} + 1\right)
$$

$$
= \left(\frac{8}{3} + \frac{6}{3}\right) - \left(\frac{1}{3} + \frac{3}{3}\right)
$$

$$
= \frac{14}{3} - \frac{4}{3} = \frac{10}{3}
$$

So, the **area is**:

$$
\frac{10}{3} \text{ square units}
$$

### 6. Distance traveled by the moving particle:

$$
x(t) = 3t^2 - 6t + 1
$$

$$
v(t) = x'(t) = \frac{d}{dt} (3t^2 - 6t + 1)
$$

$$
v(t) = 6t - 6
$$

$$
6t - 6 = 0
$$

$$
t = 1
$$

$$
\text{Distance} = \int_{0}^{1} |6t - 6| \, dt + \int_{1}^{2} |6t - 6| \, dt
$$

For $ ( t \in [0,1] )$:

$$
|6t - 6| = -6t + 6
$$

$$
\int_{0}^{1} (-6t + 6) \, dt = \left[ -3t^2 + 6t \right]_{0}^{1}
$$

$$
= (-3(1)^2 + 6(1)) - (-3(0)^2 + 6(0))
$$

$$
= (-3 + 6) - 0
$$

$$
= 3
$$

For $( t \in [1,2] )$:

$$
|6t - 6| = 6t - 6
$$

$$
\int_{1}^{2} (6t - 6) \, dt = \left[ 3t^2 - 6t \right]_{1}^{2}
$$

$$
= (3(2)^2 - 6(2)) - (3(1)^2 - 6(1))
$$

$$
= (3(4) - 12) - (3(1) - 6)
$$

$$
= (12 - 12) - (3 - 6)
$$

$$
= 0 + 3
$$

$$
= 3
$$

$$
\text{Total distance} = 3 + 3 = 6
$$
