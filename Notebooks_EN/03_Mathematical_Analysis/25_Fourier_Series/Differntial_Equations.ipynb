
# 25. Differential Equations

## 1. Solve the following first-order ordinary differential equations:

- $$ y'(x) = y(x) $$
- $$ y'(x) = \frac{1}{2y(x)} $$

## 2. Solve the first-order ordinary differential equations using the method of separation of variables for \( x \) and \( y = y(x) \):

- $$ \frac{dy}{dx} = \frac{x}{y} $$
- $$ \frac{dy}{dx} = \frac{y}{x} $$
- $$ \frac{dy}{dx} = xy $$

## 3. Solve the second-order ordinary differential equations:

- $$ y''(x) + y'(x) = 0, \quad \text{with boundary conditions } y(0) = 2 \text{ and } y'(0) = -1 $$
- $$ y''(x) - y(x) = 0, \quad \text{with boundary conditions } y(0) = 2 \text{ and } y'(0) = 0 $$
- $$ \frac{d^2y(x)}{dx^2} = -\omega^2 y(x) $$

## 4. Check if the function

$$ \psi(t, x) = A \cos(\omega t + kx) $$

is a solution of the second-order partial differential equation (the so-called "wave equation"), where

$$ v = \frac{\omega}{k} = \frac{2\pi / T}{2\pi / \lambda} $$

and

$$ \frac{\partial^2 \psi(t, x)}{\partial t^2} - v^2 \frac{\partial^2 \psi(t, x)}{\partial x^2} = 0. $$




---

## **1. Solve the first-order ODEs:**

### **(i) \( y'(x) = y(x) \)**

$$ \frac{dy}{dx} = y $$

Separation of variables:

$$ \frac{dy}{y} = dx $$

Integrating both sides:

$$ \ln |y| = x + C $$

Solving for \( y \):

$$ y = C e^x $$

---

### **(ii) $( y'(x) = \frac{1}{2y(x)} )$**

$$ \frac{dy}{dx} = \frac{1}{2y} $$

Separation of variables:

$$ 2y \, dy = dx $$

Integrating both sides:

$$ \int 2y \, dy = \int dx $$

$$ y^2 = x + C $$

$$ y = \pm\sqrt{x + C} $$

---

## **2. Solve by separation of variables:**

### **(i) $( \frac{dy}{dx} = \frac{x}{y} )$**

$$ y \, dy = x \, dx $$

Integrating:

$$ \int y \, dy = \int x \, dx $$

$$ \frac{y^2}{2} = \frac{x^2}{2} + C $$

$$ y^2 = x^2 + 2C $$

$$ y = \pm\sqrt{x^2 + 2C} $$

---

### **(ii) $( \frac{dy}{dx} = \frac{y}{x} )$**

$$ \frac{dy}{y} = \frac{dx}{x} $$

Integrating:

$$ \ln |y| = \ln |x| + C $$

$$ y = Cx $$

---

### **(iii) $( \frac{dy}{dx} = xy )\$**

$$ \frac{dy}{y} = x \, dx $$

Integrating:

$$ \ln |y| = \frac{x^2}{2} + C $$

$$ y = e^C e^{x^2/2} = C e^{x^2/2} $$

---

## **3. Solve second-order ODEs:**

### **(i) \( y''(x) + y'(x) = 0 \), with \( y(0) = 2 \), \( y'(0) = -1 \)**

Characteristic equation:

$$ r^2 + r = 0 $$

$$ r(r + 1) = 0 $$

$$ r = 0, -1 $$

General solution:

$$ y(x) = C_1 + C_2 e^{-x} $$

Using \( y(0) = 2 \):

$$ C_1 + C_2 = 2 $$

Derivative:

$$ y'(x) = -C_2 e^{-x} $$

Using \( y'(0) = -1 \):

$$ - C_2 = -1 \Rightarrow C_2 = 1 $$

Thus, \( C_1 = 1 \), and the final solution is:

$$ y(x) = 1 + e^{-x} $$

---

### **(ii) \( y''(x) - y(x) = 0 \), with \( y(0) = 2 \), \( y'(0) = 0 \)**

Characteristic equation:

$$ r^2 - 1 = 0 $$

$$ r = \pm1 $$

General solution:

$$ y(x) = C_1 e^x + C_2 e^{-x} $$

Using \( y(0) = 2 \):

$$ C_1 + C_2 = 2 $$

Derivative:

$$ y'(x) = C_1 e^x - C_2 e^{-x} $$

Using \( y'(0) = 0 \):

$$ C_1 - C_2 = 0 $$

$$ C_1 = C_2 $$

Solving:

$$ C_1 + C_1 = 2 \Rightarrow 2C_1 = 2 \Rightarrow C_1 = 1, C_2 = 1 $$

Final solution:

$$ y(x) = e^x + e^{-x} $$

---

### **(iii) $( y''(x) = -\omega^2 y(x) \$**

Characteristic equation:

$$ r^2 + \omega^2 = 0 $$

$$ r = \pm i\omega $$

General solution:

$$ y(x) = C_1 \cos(\omega x) + C_2 \sin(\omega x) $$

## **4. Verify $( \psi(t, x) = A \cos(\omega t + kx) \$ satisfies the wave equation**

Given PDE:

$$ \frac{\partial^2 \psi}{\partial t^2} - v^2 \frac{\partial^2 \psi}{\partial x^2} = 0 $$

Compute $( \frac{\partial^2 \psi}{\partial t^2} )$:

$$ \psi(t, x) = A \cos(\omega t + kx) $$

First derivative:

$$ \frac{\partial \psi}{\partial t} = -A \omega \sin(\omega t + kx) $$

Second derivative:

$$ \frac{\partial^2 \psi}{\partial t^2} = -A \omega^2 \cos(\omega t + kx) $$

Compute $( \frac{\partial^2 \psi}{\partial x^2} )\$:

First derivative:

$$ \frac{\partial \psi}{\partial x} = -A k \sin(\omega t + kx) $$

Second derivative:

$$ \frac{\partial^2 \psi}{\partial x^2} = -A k^2 \cos(\omega t + kx) $$

Check the wave equation:

$$ \frac{\partial^2 \psi}{\partial t^2} - v^2 \frac{\partial^2 \psi}{\partial x^2} = -A \omega^2 \cos(\omega t + kx) + v^2 A k^2 \cos(\omega t + kx) $$

Since $( v = \frac{\omega}{k} \$, then $ v^2 = $\frac{\omega^2}{k^2}$ \$, so:

$$ -A \omega^2 \cos(\omega t + kx) + A \omega^2 \cos(\omega t + kx) = 0 $$
