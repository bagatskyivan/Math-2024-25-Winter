### 1. **Find \( f(g(x)) \):**

We are given \( f(x) = e^x \) and \( g(x) = \ln(x) \).

$$
f(g(x)) = f(\ln(x)) = e^{\ln(x)}
$$

We know that \( e^{\ln(x)} = x \), so:

$$
f(g(x)) = x
$$

---

### 2. **Find \( g(f(x)) \):**

We are given \( f(x) = e^x \) and \( g(x) = \ln(x) \).

$$
g(f(x)) = g(e^x) = \ln(e^x)
$$

We know that \( \ln(e^x) = x \), so:

$$
g(f(x)) = x
$$

---

### Observation:

- \( f(g(x)) = x \)
- \( g(f(x)) = x \)

Thus, for these functions, the compositions \( f(g(x)) \) and \( g(f(x)) \) yield the identity function \( x \), meaning they "undo" each other. This is a special property of the exponential and logarithmic functions: they are inverses of each other.


# **4.**

To find the inverse function \( f^{-1} \), we simply swap the coordinates in each pair from the original function \( f \). The function \( f = \{(1, 7), (2, 9), (3, 11)\} \) consists of the following pairs:

- \( (1, 7) \)
- \( (2, 9) \)
- \( (3, 11) \)

Now, to find \( f^{-1} \), we swap each pair:

- \( (7, 1) \)
- \( (9, 2) \)
- \( (11, 3) \)

Thus, the inverse function is:

$$
f^{-1} = \{(7, 1), (9, 2), (11, 3)\}
$$

This means that in \( f^{-1} \), the input is the output of \( f \), and the output is the input of \( f \).


# **5.**

For a function to have an inverse, it must be **one-to-one** (meaning each input corresponds to exactly one output).

 The function \( f = \{(1, 7), (2, 7), (3, 11)\} \) is
 not one-to-one
 because both \( 1 \) and \( 2 \) map to the same value, \( 7 \).

Since a function is not one-to-one if two distinct inputs map to the same output, this function does **not** have an inverse.

In other words, the inverse function $$ f^{-1} $$ does not exist in this case.

### Finding the Inverse of \( f(x) = x - 1 \):

To find the inverse of \( f(x) = x - 1 \), follow these steps:

1. Replace \( f(x) \) with \( y \):
   $$ y = x - 1 $$

2. Swap \( x \) and \( y \) to get the inverse:
   $$ x = y - 1 $$

3. Solve for \( y \):
   $$ y = x + 1 $$

So, the inverse function is:
$$ f^{-1}(x) = x + 1 $$
