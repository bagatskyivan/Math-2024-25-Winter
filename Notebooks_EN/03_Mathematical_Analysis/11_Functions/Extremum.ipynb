### 22. Extremum

1. The profit function is $P(u) = -2u^2 + 50u - 300$, where $u$ is the number of units sold. Find the number of units that maximize profit.

2. You have 10 meters of string, and you need to use it to enclose the largest possible rectangle. Find the dimensions of the rectangle.

3. Find the extremum of $f(x) = x^2 + 3x - 5$.

4. Find the extremum of $f(x) = \frac{x^2 + 2x + 1}{x - 1}$.

---


#### **1. Maximize Profit**
The profit function is $P(u) = -2u^2 + 50u - 300$.

1.1. Find the derivative of $P(u)$:

P'(u) = $$\frac{d}{du}(-2u^2 + 50u - 300) $$= -4u + 50


1.2. Set $P'(u) = 0$ to find the critical point:
$$
-4u + 50 = 0 \implies u = 12.5
$$

1.3. Determine whether $u = 12.5$ is a maximum or minimum by checking the second derivative:
$$
P''(u) = -4
$$
Since $P''(u) < 0$, the critical point is a maximum.

**Result**: The number of units that maximizes profit is $u = 12.5$.

---

#### **2. Largest Rectangle with 10 Meters of String**
2.1. The perimeter of the rectangle is $10$, so:
$$
2(l + w) = 10 \implies l + w = 5
$$

2.2. The area of the rectangle is:
$$
A = l \cdot w
$$
Substitute $w = 5 - l$:
$$
A = l(5 - l) = 5l - l^2
$$

2.3. Maximize $A$ by finding the derivative:
$$
A'(l) = 5 - 2l
$$
Set $A'(l) = 0$:
$$
5 - 2l = 0 \implies l = 2.5
$$

2.4. Find $w$:
$$
w = 5 - l = 5 - 2.5 = 2.5
$$

**Result**: The rectangle with the largest area is a square with side lengths $2.5 \, \text{meters}$.

#### **3. Extremum of $f(x) = x^2 + 3x - 5$**
3.1. Find the derivative of $f(x)$:
$$
f'(x) = 2x + 3
$$

3.2. Set $f'(x) = 0$:
$$
2x + 3 = 0 \implies x = -\frac{3}{2}
$$

3.3. Determine whether it's a maximum or minimum by checking the second derivative:
$$
f''(x) = 2
$$
Since $f''(x) > 0$, the critical point is a minimum.

**Result**: The minimum occurs at $x = -\frac{3}{2}$, and the value of $f(x)$ is:
$$
f\left(-\frac{3}{2}\right) = \left(-\frac{3}{2}\right)^2 + 3\left(-\frac{3}{2}\right) - 5 = -\frac{29}{4}
$$

#### **4. Extremum of $f(x) = \frac{x^2 + 2x + 1}{x - 1}$**
4.1. Rewrite the numerator:
$$
x^2 + 2x + 1 = (x + 1)^2
$$
So, $f(x) = \frac{(x + 1)^2}{x - 1}$.

4.2. Find the derivative using the quotient rule:
$$
f'(x) = \frac{(2(x + 1))(x - 1) - (x + 1)^2}{(x - 1)^2}
$$

Simplify:
$$
f'(x) = \frac{(x + 1)[2(x - 1) - (x + 1)]}{(x - 1)^2} = \frac{(x + 1)(x - 3)}{(x - 1)^2}
$$

4.3. Set $f'(x) = 0$:
$$
(x + 1)(x - 3) = 0 \implies x = -1 \, \text{or} \, x = 3
$$

4.4. Check the second derivative or analyze intervals:
- $x = -1$ is undefined because $f(x)$ has a discontinuity at $x = 1$.
- $x = 3$ is a critical point.
