# 20. Limits of Real Functions

1. Compute:
$$\lim_{x \to \infty} \frac{x^3 + 2x^2}{x^4 - 3x^3}$$

2. Find:
$$\lim_{x \to 0} \frac{\sin(3x)}{2x + 1}$$

3. Find the asymptotes of the function:
$$f(x) = \frac{x^2 - 1}{x^2 + 1}, \quad g(x) = \frac{\sin(x)}{x^2 + 1}.$$

-------


### 1. Compute:
$$\lim_{x \to \infty} \frac{x^3 + 2x^2}{x^4 - 3x^3}$$

**Solution:**

We can divide both the numerator and the denominator by \( x^4 \), the highest power of \( x \) in the denominator:

$$
\lim_{x \to \infty} \frac{x^3 + 2x^2}{x^4 - 3x^3} = \lim_{x \to \infty} \frac{\frac{x^3}{x^4} + \frac{2x^2}{x^4}}{\frac{x^4}{x^4} - \frac{3x^3}{x^4}}
$$

Simplifying the terms:

$$
= \lim_{x \to \infty} \frac{\frac{1}{x} + \frac{2}{x^2}}{1 - \frac{3}{x}}
$$

As $$ x \to \infty , both  \frac{1}{x}  and  \frac{2}{x^2}  approach 0, and  \frac{3}{x} $$ also approaches 0:

$$
= \frac{0 + 0}{1 - 0} = 0
$$

So, the limit is:

$$
\lim_{x \to \infty} \frac{x^3 + 2x^2}{x^4 - 3x^3} = 0
$$

### 2. Find:
$$\lim_{x \to 0} \frac{\sin(3x)}{2x + 1}$$

**Solution:**

We use the fact that $$ \lim_{x \to 0} \frac{\sin(x)}{x} = 1 $$ and recognize that $$( 3x \to 0 ) as ( x \to 0 ):$$

$$
\lim_{x \to 0} \frac{\sin(3x)}{2x + 1}
$$

First, we observe that $$( \sin(3x) \to 0 \
 $$as
$$( x \to 0 ), and ( 2x + 1 \to 1 ) as ( x \to 0 ).$$

Thus:

$$
\lim_{x \to 0} \frac{\sin(3x)}{2x + 1} = \frac{0}{1} = 0
$$

So, the limit is:

$$
\lim_{x \to 0} \frac{\sin(3x)}{2x + 1} = 0
$$

### 3. Find the asymptotes of the functions:
$$
 ( f(x) = \frac{x^2 - 1}{x^2 + 1} )
$$
$$
 ( g(x) = \frac{\sin(x)}{x^2 + 1} )
$$

#### **For $$( f(x) = \frac{x^2 - 1}{x^2 + 1} )$$:**

**Vertical Asymptotes:**

The function \( f(x) \) has vertical asymptotes where the denominator equals 0. However, the denominator $$( x^2 + 1 ) $$never equals 0 for any real value of \( x \), so there are **no vertical asymptotes**.

**Horizontal Asymptotes:**

As $$( x \to \infty )$$, the degree of the numerator and the denominator is the same (both are degree 2). Therefore, the horizontal asymptote is the ratio of the leading coefficients:

$$
\lim_{x \to \infty} f(x) = \lim_{x \to \infty} \frac{x^2 - 1}{x^2 + 1} = \frac{1}{1} = 1
$$

Thus, the horizontal asymptote is:

$$
y = 1
$$

#### **For $$( g(x) = \frac{\sin(x)}{x^2 + 1} )$$:**

**Vertical Asymptotes:**

The function \( g(x) \) has vertical asymptotes where the denominator equals 0. Since $$ ( x^2 + 1 = 0 )$$ has no real solutions, there are **no vertical asymptotes**.

**Horizontal Asymptotes:**

As \( x \to \infty \), we can observe that $$( sin(x) ) $$oscillates between -1 and 1, while $$( x^2 + 1 \to \infty )$$. Therefore, the function $$( g(x) )$$ approaches 0 as \( x \to \infty \):

$$
\lim_{x \to \infty} g(x) = \lim_{x \to \infty} \frac{\sin(x)}{x^2 + 1} = 0
$$

Thus, the horizontal asymptote is:

$$
y = 0
$$

### Final Answers:

- For ( f(x) = $$\frac{x^2 - 1}{x^2 + 1} $$ there is a **horizontal asymptote** at ( y = 1 ) and **no vertical asymptotes**.

- For ( g(x) = $$\frac{\sin(x)}{x^2 + 1} $$ there is a **horizontal asymptote** at ( y = 0 ) and **no vertical asymptotes**.
